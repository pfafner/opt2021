```python
import numpy as np
import scipy.linalg as lg
import matplotlib.pyplot as plt
```

```python
from Iterative import Jacobi, Seidel, JOR, SOR, tridiag, pentadiag
```

## Ejemplo

```python
n = 100
A = tridiag(n, [-1,2.1,-1])
#print(A)
```

```python
np.round(A[:10,:10], 2)
```

```python
# solución exacta
xast = np.ones(n)
```

```python
b = A @ xast
print(b)
```

```python
# punto inicial
x0 = np.zeros(n)
```

```python
xJ, solJ, errJ = Jacobi(A, b, x0, tol=1e-4, maxIter=500, out=0)
```

```python
print(len(xJ))
print(np.round(xJ, 5))
```

```python
plt.figure()
plt.plot(errJ)
plt.show()
```

```python
xS, solS, errS = Seidel(A, b, x0, tol=1e-4, maxIter=500)
```

```python
print(len(xS))
print(np.round(xS, 5))
```

```python
plt.figure()
plt.plot(errS)
plt.show()
```

```python
xJor, solJor, errJor = JOR(A, b, x0, omega=0.5, tol=1e-4, maxIter=500)
```

```python
print(len(solJor))
print(np.round(xJor, 5))
```

```python
plt.figure()
plt.plot(errJor)
plt.show()
```

```python
xSor, solSor, errSor = SOR(A, b, x0, omega=0.5, tol=1e-4, maxIter=500)
```

```python
print(len(solSor))
print(np.round(xSor, 5))
```

```python
plt.figure()
plt.plot(errSor)
plt.show()
```

```python
# comparación de los métodos
plt.figure()
plt.plot(xJ, label='Jacobi')
plt.plot(xS, label='Seidel')
plt.plot(xJor, label='JOR')
plt.plot(xSor, label='SOR')
plt.legend()
plt.show()
```

```python
# cálculo de omega máximo para JOR
Dinv = np.diag(1./np.diagonal(A))
X = Dinv @ A
l, v = lg.eig(X)
wmax = 2./np.abs(l[0])
print('wmax = {}'.format(wmax))
```

```python
# cálculo de omega óptimo para SOR
D = np.diag(np.diagonal(A))
Dinv = np.diag(1./np.diagonal(A))
L = np.tril(A) - D
U = np.triu(A) - D
Tj = Dinv @ (L+U)
l, v = lg.eig(Tj)
wopt = 2./(1. + np.sqrt(1. - np.abs(l[0])**2))
print('wopt = {}'.format(wopt))
```

```python
# cambiamos a mejores parámetros
xSor, solSor, errSor = SOR(A, b, x0, omega=wopt, tol=1e-4, maxIter=500)
xJor, solJor, errJor = JOR(A, b, x0, omega=1.016, tol=1e-4, maxIter=500)
```

```python
print(len(solSor))
print(np.round(xSor, 5))
```

```python
plt.figure()
plt.plot(xJ, label='Jacobi')
plt.plot(xS, label='Seidel')
plt.plot(xJor, label='JOR')
plt.plot(xSor, label='SOR')
plt.legend()
plt.show()
```
